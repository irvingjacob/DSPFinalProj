```python
# Matplotlib is the canonical plotting library used in python
# Seaborn is built on top of Matplotlib and it makes it better!

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# What does this data look like?
iris = sns.load_dataset('iris')
iris.head()
```

```python
# How many of each species do we have?
iris["species"].value_counts()
```

```python
#iris.plot()
#iris.plot().scatter()
iris.plot(kind="scatter",x="sepal_length", y="sepal_width")
```

```python
sns.jointplot(data=iris,x="sepal_length", y="sepal_width")
```

```python
# What species is each plant from?
sns.FacetGrid(iris, hue="species", height=10) \
   .map(plt.scatter, "sepal_length", "sepal_width") \
   .add_legend()
 
```

```python
# What distribution does data follow?
iris.boxplot()
#sns.boxplot(data=iris)

# We can add some extra information to this...
#sns.stripplot(data=iris)
```

```python
# Then there is a Pandas trick for categorical boxplots side by side
iris.boxplot(by="species", figsize=(12, 6))
```

```python
sns.violinplot(data=iris)
```

```python
# sns.pairplot(data=iris)

# Can we make this better?
sns.pairplot(data=iris, hue="species")
```

```python

# One cool more sophisticated technique pandas has available is called Andrews Curves
# Andrews Curves involve using attributes of samples as coefficients for Fourier series
# and then plotting these
# https://en.wikipedia.org/wiki/Andrews_plot

from pandas.plotting import andrews_curves
andrews_curves(iris, "species")
```

```python
# https://en.wikipedia.org/wiki/Parallel_coordinates

from pandas.plotting import parallel_coordinates
parallel_coordinates(iris, "species")
```

```python

# A final multivariate visualization technique pandas has is radviz
# Which puts each feature as a point on a 2D plane, and then simulates
# having each sample attached to those points through a spring weighted
# by the relative value for that feature
# https://www.researchgate.net/profile/Ivan-Bratko/publication/225719533/figure/fig1/AS:393822853976091@1470906093826/Radviz-plot-with-two-data-examples-from-the-table-Springs-are-drawn-to-illustrate-how.png

from pandas.plotting import radviz
radviz(iris, "species")
```
